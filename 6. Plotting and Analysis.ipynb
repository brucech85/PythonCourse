```python
import plotly.express as px
import pandas as pd


import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
data=pd.read_csv('./datafiles/6705.csv')
```

```python
print(data['Litho'].unique())
data['AI']=data['Vp']*data['Density']
data['VpVs']=data['Vp']/data['Vs']
data.head()
```

```python
litho=pd.DataFrame({"LithoName":["Gas Sand","Brine Sand","Shale"],"number":[5,0,3]})
```

```python
litho
```

```python
merged=data.merge(litho,left_on='Litho', right_on='number')
merged
```

# Plotly

## Scatterplot
https://plotly.com/python/line-and-scatter/

```python
px.scatter(merged,x="AI",y="VpVs",color="LithoName")
```

```python
px.scatter(merged,x="AI",y="VpVs",color="LithoName",color_discrete_map={"Gas Sand": "red","Brine Sand":"blue","Shale":"green"})
```

```python
import plotly.express as px
df = px.data.tips()
fig = px.scatter(merged, x="AI", y="VpVs", color="Sw", facet_col="LithoName")
fig.show()
```

```python
import plotly.express as px
df = px.data.iris()
fig = px.scatter_matrix(merged,
    dimensions=["AI", "VpVs", "Density", "Vshale"],
    color="LithoName")
fig.show()
```

## Plotly GO

```python
sorted=merged.sort_values(by=['Depth'])

fig = make_subplots(rows=1, cols=3, subplot_titles=['AI','VpVs','Density'],shared_yaxes=True)
fig.add_trace(go.Scatter(x=sorted['AI'], y=sorted['Depth']),row=1, col=1)
fig.add_trace(go.Scatter(x=sorted['VpVs'], y=sorted['Depth']),row=1, col=2)
fig.add_trace(go.Scatter(x=sorted['Density'], y=sorted['Depth']),row=1, col=3)
fig.update_layout(height=800, width=600, title_text="Well A Tracks")
fig.update_yaxes(autorange="reversed")
fig.update_layout(showlegend=False)
fig.show()
```

```python
fig=make_subplots(rows=1, cols=1,subplot_titles=['VpVs'],shared_yaxes=True)
fig.add_trace(go.Bar(y=sorted['Depth'].values,x=sorted['VpVs'].values,name="VpVs", orientation='h',offset=1, marker=dict(
        color=sorted['VpVs'],line=dict(color='rgba(128, 128, 128,1)', width=0.25))),row=1,col=1)
fig.update_layout(barmode='stack',height=800, width=600)
fig.update_yaxes(autorange="reversed")
fig.update_layout(showlegend=True)
fig.show()
```
