# Slicing a Data Frame

```python
print(data['Litho'].unique())

sand=data.loc[(data['Litho']==5.0)&(data['Vp']>2800)]
sand.describe()
```

# Output a dataframe

```python
#sand.to_csv('./mycsv')

#sand.to_excel(''./myxls.xls',sheet_name="Data")

sand[['Vp','Vs','Density']].to_csv('./datafiles/mydata.csv',sep='\t')

#sand[['Vp','Vs','Density']].to_csv('./mydata.csv',sep='\t',index=False,header=False)
```

# Aggregation and Merge

```python
data.groupby(['Litho']).mean()
```

```python
litho_meta=pd.DataFrame({'Litho':[0.0,3.0,5.0],'Name':['Undef','Sand','Shale']})
litho_meta
```

```python
data.merge(litho_meta, left_on='Litho', right_on='Litho')
```

# Apply functions on Dataframe

```python
#single paramater

def scale_peff(x):
    y=x*0.1356
    return y

data['ScalledPeff']=data['Peff'].apply(scale_peff)
```

```python
data.head()
```

```python
#multi paramater pass in a dataframe row

def calc_ai(row):
    ai=row['Vp']*row['Density']
    return ai

data['AI']=data.apply(calc_ai, axis=1)
    
```
