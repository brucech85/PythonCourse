```python
import pandas as pd
data=pd.read_csv(r'./datafiles/6705.csv')
#print(type(data))

data.sample(10)
```

```python
data['VpVs Ratio']=data['Vp']/data['Vs']

def calc_ai(row):
    ai=row['Vp']*row['Density']
    return ai

data['AI']=data.apply(calc_ai, axis=1)
    
```

```python
litho_meta=pd.DataFrame({'Litho':[0.0,3.0,5.0],'Name':['Undef','Sand','Shale']})
litho_meta
data_new=data.merge(litho_meta, left_on='Litho', right_on='Litho')
data_new.head()
```

```python
import plotly.express as px
fig = px.parallel_coordinates(data_new[['VpVs Ratio','AI','Vshale','Peff','Litho']], color="Litho", color_continuous_scale=px.colors.sequential.Inferno)
fig.show()
```

```python
fig = px.scatter(data_new[['VpVs Ratio','AI','Vshale','Peff','Name']], x="AI", y="VpVs Ratio",color="Peff", facet_col="Name")
fig.show()
```

# Machine Learning

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
features=['VpVs Ratio','AI','Peff']
target=['Name']

X_train, X_test, y_train, y_test = train_test_split(data_new[features], data_new[target], test_size=0.10, random_state=42)
print(len(y_test))


clf = SVC(gamma='auto')



clf.fit(X_train, y_train.values.ravel())
```

```python
print(clf.predict(X_test))
```

```python
from sklearn.metrics import accuracy_score

y_pred = clf.predict(X_test)
y_true = y_test

accuracy_score(y_true, y_pred, normalize=True)


```
