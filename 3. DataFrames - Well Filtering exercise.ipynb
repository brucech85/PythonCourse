```python
import pandas as pd
import numpy as np
from pandas import ExcelWriter 
```

```python
fields=['BALDER','COD','TOR','FRAM','GOLIAT','EKOFISK','GRANE']
```

```python
prod_wells=pd.read_csv(r"./datafiles/wellbore_development_all.csv", sep=",")
print(prod_wells.columns)
prod_wells.head()
```

```python
wells_in_area=prod_wells.loc[(prod_wells['wlbField'].isin(fields))]
```

```python
wells_in_area.describe()
```

```python
group=wells_in_area[['wlbField','wlbWell']].groupby(by=["wlbField"]).count()
group

```

```python
w = ExcelWriter('Field_Wells.xlsx')
for a in wells_in_area['wlbField'].unique():
    df=wells_in_area.loc[(wells_in_area['wlbField']==a)]
    df.to_excel(w, sheet_name=a)
group.to_excel(w, sheet_name="Summary")
w.save()
```
