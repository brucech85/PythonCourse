# Calculation on rows

```python
data['TotalTime']=data['Completed date']-data['Entered date']
data['TTnum']=data['TotalTime'].dt.days
data.head()
#data['TotalTimeAbs']=data['TotalTime'].apply(abs)
#filtered.append('TotalTimeAbs')
```

# Aggregation

```python
f2=['Drilling operator','Purpose','TTnum']
grouped=data.groupby(by=['Drilling operator','Purpose','Content']).mean()
grouped.reset_index(inplace=True)
#print(grouped.dtypes)
grouped.head()
```

```python
import plotly.express as px
fig = px.bar(grouped, x="Drilling operator", y="TTnum",color='Purpose',width=800, height=800)
fig.show()
```

```python
fig = px.sunburst(grouped.sample(25), path=['Drilling operator','Purpose', 'Content'], values='TTnum')
fig.show()
```
