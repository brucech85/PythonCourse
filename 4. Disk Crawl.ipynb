```python
paths = ["Python38","HP","Google"]



for b in paths:

  path = r'C:\\Program Files\\'+b
  file_object = open('./'+b+"files"+'.txt', 'a')

        
  print(os.listdir(path))
 
  #r=root, d=directories, f = files
  for r, d, f in os.walk(path):
    if len(d) != 0:
      #print(d)
      #print(r)
      if len(f) !=0:
        for a in f:
            if a.endswith(".exe"):
              print(r+"\\"+a+'\n')
              file_object.write(r+"\\"+a+'\n')
                
  file_object.close
        

    
```
